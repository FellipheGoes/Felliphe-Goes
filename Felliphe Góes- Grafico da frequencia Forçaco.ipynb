```python
import scipy as sp
import matplotlib.pyplot as plt
from ipywidgets import interact
%matplotlib inline 
```

```python
wn=3.1415
F0=1
wf=sp.linspace(0,2*3.1415,1000)
```

```python
def X(epi):
    X=F0/((wn**2-wf**2)**2 + (2*epi*wn*wf)**2)
    plt.plot(wf/wn,X)
```

```python
def Teta(epi):
    Teta=sp.arctan(2*epi*wn*wf/(wn**2-wf**2))
    plt.plot(wf/wn,Teta)
```

```python
interact(X  , epi=(0,1,.01))
```

```python
interact(Teta  , epi=(0,1,.01))
```
