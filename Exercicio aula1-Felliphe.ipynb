```python
t=sp.linspace(0,25,1000)
```

```python
import scipy as sp
import matplotlib.pyplot as plt
from ipywidgets import interact
%matplotlib inline 
```

```python
v0=0
x0=.1
wn=2*2*sp.pi
#wf=2.1*2*sp.pi
f=1
```

```python
def func(wf):
    x = (v0/wn)*sp.sin(wn*t)+(x0-f/(wn**2-wf**2))*sp.cos(wn*t)+(f/(wn**2-wf**2))*sp.cos(wf*t)
    plt.plot(t,x)
```

```python
interact(func , wf=1.5)
```
