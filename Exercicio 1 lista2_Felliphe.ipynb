```python
import scipy as sp
from scipy.integrate import odeint
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def pendx(x,t):
        F0=150
        c=100
        k=2000
        m=100
        w=10
        [x1,x2]=x
        dxdy=[x2, m*(w**2)*sp.cos(2*sp.pi*w*t)/m-(c/m)*x2-(k/m)*x1]
        return dxdy
        
```

```python
def pendy(y,t):
        F0=150
        c=100
        k=2000
        m=100
        w=10
        [y1,y2]=y
        dxdy=[y2, m*(w**2)*sp.sin(sp.pi*2*w*t)/m-(c/m)*y2-(k/m)*y1]
        return dxdy
        
```

```python
x0=0
v0=0

z=[x0, v0]
tf=10
t=sp.linspace(0,tf,10000)
```

```python
solx=odeint(pendx,z,t)
```

```python
soly=odeint(pendy,z,t)
```

```python
plt.plot(t,solx[:,0], label='x1(t)')
plt.legend()
plt.xlabel('t')
```

```python
plt.plot(t,soly[:,0], label='y1(t)')
plt.legend()
plt.xlabel('t')
```

```python
plt.plot(solx[:,0],soly[:,0])

plt.xlabel('x(t)')
plt.ylabel('y(t)')
```
